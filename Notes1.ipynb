### Constant Velocity

Assume you have a cart on a frictionless track. At time t = 0 seconds, the cart is at a position x = 0 m. The cart is moving with a constant velocity of 0.45 m/s (in the x-direction). 
- Where is the cart at time t = 1.5 seconds? 
- What about at 0.25 seconds?
- Assume the cart continues to travel for another 0.25 seconds (a total of 0.5 seconds) where is it now?
- How many steps of 0.25 seconds would you need to calculate in order to determine the position at 1.5 seconds? 
- What if you used a time step size of 0.1 seconds?

In that final case you could imagine filling in a table that looked like:

| time (s) | x position (m) |
| :--- | :--- |
| 0 | 0 |
| 0.1 | 0.045 | 
| 0.2 | ... |

If you know the position (x value) for an object at a particular time and you also know the average velocity, you can find the position at the next time. Suppose I have the third position and I want want to find the 4th position. This would be the formula:
$$x_4 = x_3 + v \Delta t$$ where $\Delta t$ is the time step size.

Clearly, you could do the calculations above for any number of steps. It might get crazy for a long time or a small time step, but you could theoretically do it. Using our python skills we can do this.

``` python

import matplotlib.pyplot as plt
x = 0 #this is the inital position
t = 0 #this is the inital time
v = 0.45 #this is the velocity
dt = 0.1 #this is the time-step

X=[x] #these will be the arrays of all x and time
T=[t]

while t<1.5:
    x=x+v*dt #this calculates the next position
    t=t+dt #this calculates the next time
    X.append(x) #this 'appends' the latest value of x and t to the arrays X and T
    T.append(t)
    
    
print('at time t=',t,'our x value is',x)
#notice the print statement is not indented so is not part of the while loop.

```

````python

plt.plot(T,X,'c-+') #this plots the array T on the x-axis and array X on the y-axis. 
plt.title('Position-time graph') #you can work out the next 3 lines
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show()
````

### Constant acceleration

A cart starts at x = 0 meters with a velocity of 0.45 m/s at a time of t = 0 seconds. If the cart has an acceleration of 0.2 m/s<sup>2</sup>, where will the cart be after 1.5 seconds?

Let’s do this acceleration problem by taking small time steps of 0.25 seconds. You can still use the position update formula – but what about the velocity? From the definition of acceleration, I get the following for the velocity update formula: $v_2 = v_1 + a\Delta t$

So in each time step you can calculate the new velocity based on the acceleration, the old velocity and the time step. What about the change in position? You can still use the position update formula from above – but we need to make a choice. Since we might not know the average velocity, we will just use the final velocity instead. Here is the order of calculations for each time step:
 - Use the velocity update formula to find the new velocity
 - Use the position update formula (with the new velocity) to find the new position
 - Update time.
 
Again we can make a table of these updates
 
| time (s) |v (m/s)| x position (m) |
| :-: | :-: | :-:|
| 0 |  0.45| 0 |
| 0.25 | ... | ... |
| 0.5 | ... | ... |

``` python

import matplotlib.pyplot as plt
x = 0 #this is the inital position
t = 0 #this is the inital time
v = 0.45 #this is the inital velocity
a=0.2 #this is the acceleration

dt = 0.1 #this is the time-step

X=[x] #these will be the arrays of all x, time and velocity
T=[t]
V=[v]

while t<1.5:
    v=v+a*dt #this calculates the next velocity
    x=x+v*dt #this calculates the next position
    t=t+dt #this calculates the next time
    X.append(x) #this 'appends' the latest value of x and t to the arrays X, T and V
    T.append(t)
    V.append(v)
    
    
print('at time t=',t,'our x value is',x)
#notice the print statement is not indented so is not part of the while loop.
```


at time t=1.5 our x value is 0.915

````python
plt.plot(T,X,'c-+') #this plots the array T on the x-axis and array X on the y-axis. 
plt.title('Position-time graph') #you can work out the next 3 lines
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show()

plt.plot(T,V,'m-+')
plt.title('Velocity-time graph')
plt.xlabel('time (s)')
plt.ylabel('v (m/s)')
plt.show()
````
