## 10 dimensional hypersphere

Although physical space has only three dimensions, mathematically we sometimes need to think in
more then three variables. It is easy to do higher dimensional integrals using the Monte Carlo method.
In this problem we will calculate the volume of a 10-dimensional sphere (called a _hypersphere_) encased
in a 10-dimensional _hypercube_ of side 2 units.

The formula for the unit hypersphere is given by

$$
x_1^2+x_2^2+x_3^2+...+x_{10}^2 =1
$$

We start by generating random numbers $x_1, x_2, x_3, . . . x_{10}$ in the hypercube $(−1 ≤ x_1, x_2, x_3, . . . x_{10} ≤ 1)$.
Count the ratio, $R$, of these random sets that lie inside the sphere. Another way of saying this is count
the ratio of these random sets that satisfy:

$$
x_1^2+x_2^2+x_3^2+...+x_{10}^2 \le1
$$


Since the accuracy of the Monte Carlo method grows as $\frac{1}{\sqrt{N}}$ (where $N$ is the number of random sets of
$x_{1}, x_{2}, x_{3}, . . . x_{10}$ you choose) you will need to choose a large $N$ for your calculation. How do you check it is large enough?

## Calculate the mass, centre of mass and moment of inertia of a complicated shape


We will use the Monte Carlo method as in the previous exercise to determine the mass of a complicated
shape with a constant density, the centre of mass of this shape and its moment of inertia.

The mass of an object is defined as:
$$
\int  \rho \,dx \,dy \,dz
$$
the centres of mass $(x_{cm}, y_{cm}, z_{cm})$ are defined
 as:

$$
x_{cm}=\frac{\int x \rho \,dx\,dy,\,dz}{\int \rho \,dx\,dy,\,dz} \qquad 
y_{cm}=\frac{\int y \rho \,dx\,dy,\,dz}{\int \rho \,dx\,dy,\,dz} \qquad 
z_{cm}=\frac{\int z \rho \,dx\,dy,\,dz}{\int \rho \,dx\,dy,\,dz}
$$
and finally the momemts of inertia are:
$$
I_{xx} = \int (y^2 + z^2) \rho \,dx\,dy\,dz \qquad 
I_{yy} = \int (x^2 + z^2) \rho \,dx\,dy\,dz \qquad 
I_{zz} = \int (x^2 + y^2) \rho \,dx\,dy\,dz
$$
